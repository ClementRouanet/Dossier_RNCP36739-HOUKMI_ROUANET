## **Dataset columns:**
  **•	age:** The person’s age in years \
\
  **•	sex:** The person’s sex (0 = female, 1 = male) \
\
  **•	cp:** chest pain type \
      — Value 0: asymptomatic \
      — Value 1: atypical angina \
      — Value 2: non-anginal pain \
      — Value 3: typical angina \
\
  **•	trestbps:** The person’s resting blood pressure (mm Hg on admission to the hospital) \
\
  **•	chol:** The person’s cholesterol measurement in mg/dl \
\
  **•	fbs:** The person’s fasting blood sugar (> 120 mg/dl, 1 = true; 0 = false) \
\
  **•	restecg:** resting electrocardiographic results \
      — Value 0: showing probable or definite left ventricular hypertrophy by Estes’ criteria \
      — Value 1: normal \
      — Value 2: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV) \
\
  **•	thalach:** The person’s maximum heart rate achieved \
\
  **•	exang:** Exercise induced angina (1 = yes; 0 = no) \
\
  **•	oldpeak:** ST depression induced by exercise relative to rest (‘ST’ relates to positions on the ECG plot) \
\
  **•	slope:** the slope of the peak exercise ST segment \
      0: downsloping ; 1: flat ; 2: upsloping \
\
  **•	ca:** The number of major vessels (0–3) \
\
  **•	thal:** A blood disorder called thalassemia \
      — Value 0: NULL \
      — Value 1: fixed defect (no blood flow in some part of the heart) \
      — Value 2: normal blood flow \
      — Value 3: reversible defect (a blood flow is observed but it is not normal) \
\
  **•	target:** Heart disease (0 = no, 1 = yes)

# Nettoyage des données avec des valeurs nulles

```python
# pourcentage de valeurs nulles par champ

100 * df.isnull().sum() / len(df)
```

```python
# Prend un dataframe en entrée et renvoie les champs ayant des valeurs nulles

def percent_missing(df):
    percent_nan = 100* df.isnull().sum() / len(df)
    percent_nan = percent_nan[percent_nan>0].sort_values()
    return percent_nan
```

```python
percent_nan = percent_missing(df)
```

```python
plot = sns.barplot(x=percent_nan.index,y=percent_nan)
plot.set_ylabel("Pourcentages")
plot.set_title("Pourcentage de valeurs nulles par champ")

for p in plot.patches:
    plot.annotate(f'{p.get_height():.1f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```

# Preprocessing

```python
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay, accuracy_score, precision_score, recall_score
```

```python
# Séparation du dataframe en deux : la target et le reste

X = df.drop('target',axis=1)
y = df['target']
```

```python
# Séparation de X et y en deux partie : test et train

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
# Standardisation des caractéristiques

scaler = StandardScaler()

scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

## SVM (Support Vector Machine)

```python
from sklearn.svm import SVC
```

```python
svc = SVC(class_weight='balanced')
```

```python
# Recherche des meilleurs paramètres pour le modèle

param_grid = {'C':[0.001,0.01,0.1,0.5,1], 'gamma':['scale','auto']}
best_svm = GridSearchCV(svc, param_grid)

best_svm.fit(scaled_X_train, y_train)

best_svm.best_params_
```

```python
# Prédiction avec le modèle entraîné

y_pred = best_svm.predict(scaled_X_test)
```

```python
# Matrice de confusion

cm = confusion_matrix(y_test,y_pred)

cm_plt = ConfusionMatrixDisplay(confusion_matrix=cm)
cm_plt.plot(cmap='Blues', values_format='d')
plt.title("Matrice de confusion du modèle SVM")
plt.xlabel('Prédictions')
plt.ylabel('Vraies étiquettes')

plt.show()
```

```python
print(classification_report(y_test,y_pred))
```

# KNN (Nearest Neighbour classification)

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
k_values = list(range(1,30))
param_grid = {'n_neighbors': k_values}

best_knn = GridSearchCV(knn,param_grid,cv=5,scoring='accuracy')

best_knn.fit(scaled_X_train,y_train)

best_knn.best_estimator_.get_params()
```

```python
# Obtenir les scores
scores = best_knn.cv_results_['mean_test_score']

# Trouver l'indice du score maximal
max_index = np.argmax(scores)

# Obtenir les valeurs correspondantes de k
k_max = k_values[max_index]
score_max = scores[max_index]

plt.plot(k_values,scores,'o-')
plt.scatter(k_max, score_max, color='red', marker='o', zorder=5, s=70)
plt.text(k_max, score_max, f'   K={k_max}', color='red', verticalalignment='bottom')
plt.xlabel("K")
plt.ylabel("Accuracy")
plt.title("Accuracy du modèle KNN en fonction du K")

plt.show()
```

```python
# Prédiction avec le modèle entraîné

y_pred = best_knn.predict(scaled_X_test)
```

```python
#Matrice de confusion

cm = confusion_matrix(y_test,y_pred)

cm_plt = ConfusionMatrixDisplay(confusion_matrix=cm)
cm_plt.plot(cmap='Blues', values_format='d')
plt.title("Matrice de confusion du modèle KNN")
plt.xlabel('Prédictions')
plt.ylabel('Vraies étiquettes')

plt.show()
```

```python
print(classification_report(y_test,y_pred))
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from scipy.stats import randint
```

```python
# Création a random forest classifier
rf = RandomForestClassifier()
```

```python
# Utilisation de la recherche aléatoire pour trouver les meilleurs hyperparamètres

param_dist = {'n_estimators': randint(100,1000),
              'max_depth': randint(1,20)}


rand_search = RandomizedSearchCV(rf,
                                 param_distributions = param_dist,
                                 n_iter=5,
                                 cv=5)

rand_search.fit(scaled_X_train, y_train)
```

```python
# Variable pour le meilleur modèle
best_rf = rand_search.best_estimator_

print('Meilleurs hyperparamètres:',  rand_search.best_params_)
```

```python
# Prédiction avec le modèle entraîné
y_pred = best_rf.predict(scaled_X_test)
```

```python
#Matrice de confusion

cm = confusion_matrix(y_test,y_pred)

cm_plt = ConfusionMatrixDisplay(confusion_matrix=cm)
cm_plt.plot(cmap='Blues', values_format='d')
plt.title("Matrice de confusion du modèle random Forest")
plt.xlabel('Prédictions')
plt.ylabel('Vraies étiquettes')

plt.show()
```

```python
print(classification_report(y_test,y_pred))
```

# Comparaisons des modèles

```python
import warnings

# Ignorer les avertissements spécifiques liés à FutureWarning
#warnings.filterwarnings("ignore", category=FutureWarning, module='sklearn')

models = [
    [best_svm, "SVM"],
    [best_knn, "KNN"],
    [best_rf, "RF"]
]

accurracies = []
scores = []

# Evaluer chaque modèle
for model in models:
    accuracy = model[0].score(scaled_X_test, y_test)
    accurracies.append(accuracy)
    print(f"Accuracy for {model[1]}: {accuracy:.4f}")

print('\n')

# Comparaison avec validation croisée
for model in models:
    score = cross_val_score(model[0], X, y, cv=5)
    scores.append(score.mean())
    print(f"Mean accuracy for {model[1]}: {score.mean():.4f}")

#warnings.resetwarnings()
```

```python
plot = sns.barplot(x=["SVM", "KNN", "RF"], y=accurracies)
plot.set_title("Accurracy pour chaque modèle")

for p in plot.patches:
    plot.annotate(f'{p.get_height():.4f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```

```python
plot = sns.barplot(x=["SVM", "KNN", "RF"], y=scores)
plot.set_title("Score moyen de la validation croisée pour chaque modèle")

for p in plot.patches:
    plot.annotate(f'{p.get_height():.4f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.show()
```
